# Introducción

# Idea

Descomposición LU.

Tanto para resolver el sistemas como para calcular la inversa vamos a usar la descomposición *LU* .
Sea $A \in \mathbb{R}^{n \times n}$ a la cual le queremos calcular la descomposición *LU* . Para ello, lo calcularemos en
bloques en vez del método tradicional de triangulación.
Supongamos que *A* tiene descomposición *LU* ; es decir podemos descomponerla en *L*, *U* $\in \mathbb{R}^{n \times n}$
escribiendo la descomposición en bloques de la forma:

A = LU

con L21 U12 un producto de vector columna por vector fila (que resulta en una matriz). Aquı́ lleg-
amos a una ecuación donde nuestras incógnitas son las dos matrices L22 (triangular inferior con
unos en la diagonal) y U22 (triangular superior), es decir, una descomposción LU . Si resolvemos
(recursivamente) la descomposición LU de A22 − L21 U12 , podemos obtener los dos bloques que nos
faltan.


\begin{equation*}
\begin{pmatrix}
a_{11} & A_{12}\\
A_{21} & A_{22} \\
\end{pmatrix}
=
\begin{pmatrix}
l_{11} & 0 \\
L_{21} & L_{22} \\
\end{pmatrix}
\begin{pmatrix}
u_{11} & U_{12}\\
0 & U_{22}\\
\end{pmatrix}
\end{equation*}

\begin{equation*}
\begin{pmatrix}
a_{11} & A_{12}\\
A_{21} & A_{22} \\
\end{pmatrix}
=
\begin{pmatrix}
l_{11} & U_{12} \\
L_{21} u_{11} & L_{21} U_{12} + L_{22} U_{22} \\
\end{pmatrix}
\end{equation*}

Despejando, nos queda: $A_{22} - L_{21} U_{12} = L_{22} U_{22}$

# Ejercicios 

Importamos librerias

```python
import numpy as np
from scipy.linalg import solve_triangular
import matplotlib.pyplot as plt
from numpy.linalg import LinAlgError
from progress.bar import Bar
```

Cargamos funciones auxiliares que usaremos a lo largo del notebook

```python
def esMatrizCuadrada(A):
    return A.shape[0] == A.shape[1]

def matriz_H(n, p=np.float64 ):
    H = np.zeros((n,n),dtype=p)
    for i in range(n):
        for j in range(n):
            H[i][j] = 1/(i+j+1)
    return H


def norma2(v, l, r):
    if l+1 < r:
        mid = (l+r)//2
        n_left = norma2(v, l, mid)
        n_right = norma2(v, mid, r)
        return n_left + n_right
    else:
        return v[l]*v[l]

def norma2Vectorial(v):
    vc = v.copy()
    if vc.shape[0]>1 : # Corregir est condición
        vc = np.transpose(vc)[0]
    norma = norma2(vc,l=0,r=v.shape[0])
    
    return np.sqrt(norma)

def normaFrobenius(A):
    suma_cuadrados = np.sum(np.square(A))
    normaF = np.sqrt(suma_cuadrados)
    return normaF

#función para generar matrices dada una condicion externa
def generarMatriz(fil=2, col=2, metodo='random', radio=1, p=np.float64):
    if metodo == 'random':
        return (2*radio * np.random.rand(fil,col) - radio).astype(p)
    elif metodo == 'H' and fil == col:
        return matriz_H(fil,p=p)
    
# si n = 0, quiere decir que la prediccion es exacta, sin embargo para propositos practicos de este trabajo es inutil esta información
# ádemas que retorna un valor menos infinito, que anula automaticamente los errores cometidos en otras predicciones durante el ciclo.
def errorRelativo(M,v_x, v_xp, v_b, funcion_norma,  condicion=True):
    if condicion:
        n = funcion_norma(M@v_xp-v_b)
        d = funcion_norma(v_b)
    else:
        n = funcion_norma(v_xp-v_x)
        d = funcion_norma(v_x)
    if n == 0 or d == 0:
        return 0
    else:
        return np.log(n/d)

def permutacion(A, vector_P, index=0):
    # dada una columna col_k(A), buscamos el indice i del valor absoluto mas alto de dicha columna, luego hacemos swap
    k = np.argmax(np.abs(A[index:,index:index+1]))
    # swapiamos
    fila_i = np.array(A[index])
    if  k == 0:
        print("Fallo!, error en matriz")
        return
    k += index
    A[index] = A[k]
    A[k] = fila_i
    vector_P[index], vector_P[k] = vector_P[k], vector_P[index]
```

## Descomposición LU
Dada una matriz $A \in \mathbb{R}^{n \times n}$ la función **descompLU** retorna dos matrices L y U de la desocomposición $A=LU$

```python
def descompLU(A, verbose=False):
    Ac = A.copy()
    n = A.shape[0]-1
    i = 0
    # Descomponemos y operamos
    while i<n:
        a_11 = Ac[i][i]
        if a_11 == 0: 
            print("cero en diagonal de matriz")
            return Ac,A
            break
        U_12 = Ac[i:i+1,i+1:]        
        L_21 = np.divide(Ac[i+1:,i:i+1],a_11)
        Ac[i+1:,i:i+1] = L_21
        Ac_i = np.subtract( Ac[i+1:,i+1:], L_21@U_12)
        Ac[i+1:,i+1:] = Ac_i
        i+=1
        
    L = np.tril(Ac,-1) + np.eye(A.shape[0], dtype=A.dtype) 
    U = np.triu(Ac)
    if verbose:
        print("Matriz L \n", L)
        print("Matriz U \n", U)        
    return L,U
```

# Descomposición PA = LU
Dada una matriz $A \in \mathbb{R}^{n \times n}$ la función **descompPLU** retorna tres matrices p, L y U de la desocomposición $PA=LU$

```python
def descompPLU(A, verbose=False):
    Ac = A.copy()
    n = A.shape[0]-1
    i = 0
    v_p = np.arange(n+1)
    # Descomponemos y operamos
    while i<n:        
        if Ac[i][i] == 0: 
            permutacion(Ac, vector_P=v_p, index=i)
        a_11 = Ac[i][i]
        U_12 = Ac[i:i+1,i+1:]        
        L_21 = np.divide(Ac[i+1:,i:i+1],a_11)
        Ac[i+1:,i:i+1] = L_21
        Ac_i = np.subtract( Ac[i+1:,i+1:], L_21@U_12)
        Ac[i+1:,i+1:] = Ac_i
        i+=1
    P = np.eye(n+1, dtype=A.dtype)
    L = np.tril(Ac,-1) + P 
    U = np.triu(Ac)
    P = P[v_p]
    if verbose:
        print("Matriz P \n", P)
        print("Matriz L \n", L)
        print("Matriz U \n", U)        
    return P,L,U
```

## Uso de prueba

```python
A = np.array([[0,0,2,4],[0,2,4,3],[5,2,1,1],[2,3,1,1]]) 
P,L,U = descompPLU(A)
A = np.array([[5,2,1,1],[0,2,4,3],[0,0,2,4],[2,3,1,1]]) 
B,C = descompLU(A)
print('B=L? ' , 'Si!' if np.allclose(np.linalg.norm(B - L, 1), 0) else 'No!')
print('C=U? ' , 'Si!' if np.allclose(np.linalg.norm(C - U, 1), 0) else 'No!')
```

## Resolver sistemas

### Resolver sistema Ax=b con A=LU

La función **resolverLU** recibe una matriz $A \in \mathbb{R}^{n \times n}$ y un vector $b \in \mathbb{R}^n$, retorna *x* la solución del sistema $Ax=b$ calculando primero la descomposición *LU*, por lo que tendremos:\
$Ax=b$, donde $A = LU$ entonces $LUx=b $ \
Obtenenos entonces dos sistemas a resolver:

$\left\{ \begin{array}{lcc}
    Ly=b \\
    Ux=y
    \end{array}\right.$


```python
# LUx = b => Ly=b, Ux=y
def resolverLU(A, b):
    n = A.shape[0]   
    L,U = descompLU(A, verbose=False)
    y = solve_triangular(L, b, lower=True) # Ly = b
    x = solve_triangular(U, y) # Ux = y
    return x
```

### Resolver sistema Ax=b con PA=LU

La función **resolverPLU** recibe una matriz $A \in \mathbb{R}^{n \times n}$ y un vector $b \in \mathbb{R}^n$, retorna *x* la solución del sistema $Ax=b$ calculando primero la descomposición *LU*, por lo que tendremos:\
$Ax=b$, donde $PA = LU$ entonces $LUx=b $ \
Obtenenos entonces dos sistemas a resolver:

$\left\{ \begin{array}{lcc}
    Ly=b \\
    Ux=y
    \end{array}\right.$

```python
def resolverPLU(A, b):
    n = A.shape[0]
    if not esMatrizCuadrada(A) or b.shape[0] != n:
        print("Matriz y/o vector no válidos")
        return 0    
    P,L,U = descompPLU(A, verbose=False)
    y = solve_triangular(L, b, lower=True).astype(A.dtype) # Ly = b
    x = solve_triangular(U, y).astype(A.dtype) # Ux = y
    return x
```

## Cálculo de la inversa

La función **resolverLU** recibe una matriz $A \in \mathbb{R}^{n \times n}$ inversible, esta función retorna la inversa de A realizando los siguientes pasos:
 - Calcula la descomposición LU de A
 - Para cada vector $e_i, 1 \leq i \leq n$ de la base canónica de $\mathbb{R}^n$, resuelve los sistemas \
 $\left\{ \begin{array}{lcc}
    Ly=e_i \\
    Ux=y
    \end{array}\right.$
 

```python
def inversa (A) :
    n = A.shape[0]
    i = 0
    try: 
        L,U = descompLU(A)
    except ValueError:
        P,L,U = descompPLU(A)
    if np.allclose(np.linalg.norm(A - U, 1), 0): # si descompLU no funciona, probamos con descompPLU
        P,L,U = descompPLU(A)
    inversa = []
    try:
        while i < n:        
            e_i = np.zeros(n)
            e_i[i] = 1
            y = solve_triangular(L, e_i, lower=True).astype(A.dtype)
            x = solve_triangular(U, y, lower=False).astype(A.dtype)
            inversa.append(x)
            i+=1
    except LinAlgError as e:
        print("Error de algebra lineal : ", e.args[0])
    matriz = np.array(inversa)    
    inversa = np.transpose(matriz)
    return inversa
```

# Experimentos

Los experimentos que realizaremos tienen la siguiente estructura:

Para i desde n hasta m, $m,n \in \mathbb{N}$ se realizaran los siguientes experimentos.
1. Repetir k veces:
    - Tomar una matriz $A \in \mathbb{R}^{n \times n}$ de números aleatorios en $[-1,1)$  y un vector $b \in \mathbb{R}^n$ de números aleatorios en $[-1,1)$. Definir $b = Ax$.
    - Calcular la solución $x_1$ del sistema $Ax=b$ mediante la función resolverLU.
    - Calcular el logaritmo natural del error relativo $\frac{\|Ax-b \|_2}{\|b\|_2}$ o $\frac{\|x_1-x \|_2}{\|x\|_2}$
    - Calcular la solución $x_2$ del sistema $Ax=b$ mediante la fórmula $x_2 = A^{-1}b$, utilizando la función inversa implementada anteriormente.
    - Calcular el logaritmo natural del error relativo $ \frac{\|Ax-b \|_2}{\|b\|_2}$ o $\frac{\|x_2-x \|_2}{\|x\|_2}$
2. Grabar en un vector v1 los errores cometidos por el primer método.
3. Grabar en un vector v2 los errores cometidos por el segundo método.

Por lo que la componente de cualquier vector de dimensión m tendrá la forma:

$v_i =\left\{ \begin{array}{lcc}
    \sum_{j=0}^k{Er(j)} & si & n \leq i \leq m \\
    0 & si & i < n
    \end{array}\right.$
    
donde $Er(k) = ln(\frac{\|A_k*(x_p)_k-b_k \|_2}{\|b_k\|_2})$ o $ Er(j) = ln(\frac{\|(x_p)_k-x_k \|_2}{\|x_k\|_2})$, $A \in \mathbb{R}^{k \times k}$ y $x, (x_p), b \in \mathbb{R}^k$,


En cada una de las k repeticiones se tomara una nueva matriz aleatoria.
Graficaremos los errores cometidos por ambos métodos en un solo gráfico.
Para tener mas claridad a la hora de realizar los experimentos, creamos tres funciones, las cuales automatizan el proceso de experimentacion antes descripto y nos otroga facilidad en su lectura y cambiar parametros por cada experimento.

```python

def perfomance(dict_func, rango_de_iteracion=(2,10), ciclos=10, tipo_de_matriz='random', tipo_num=np.float64 ,er_b=True, verbose=True):
    v1 = np.zeros(rango_de_iteracion[1]+1, dtype=tipo_num)
    v2 = np.zeros(rango_de_iteracion[1]+1, dtype=tipo_num)
    for i in range(rango_de_iteracion[0],rango_de_iteracion[1]+1):
        e1,e2 = tipo_num(0), tipo_num(0)
        for j in range(ciclos):
            A = generarMatriz(fil=i,col=i,metodo=tipo_de_matriz, p=tipo_num)
            x = (2 * np.random.rand(i,1) - 1).astype(tipo_num)
            b = A@x
            A_i = dict_func['inv'](A)
            xp = dict_func['solve'](A,b)
            e = errorRelativo(M=A, v_x=x, v_xp=xp, v_b=b, funcion_norma=dict_func['norma'])
            e1 += e 
            xp = A_i@b
            e = errorRelativo(M=A, v_x=x, v_xp=xp, v_b=b, funcion_norma=dict_func['norma'], condicion=False)
            e2 += e        
        v1[i] = e1
        v2[i] = e2
    if verbose:
        print("Finalizó con exito")
              
    return v1, v2


def plotearVectores(dicc_v1, dicc_v2, rotulos, start=10):
    if len(dicc_v1['vector']) != len(dicc_v2['vector']):
        print("Parametros invalidos")
        return 
    n = np.arange(start,len(dicc_v1['vector']))
    plt.plot(n,dicc_v1['vector'][start:], label=dicc_v1['leyenda'])
    plt.plot(n,dicc_v2['vector'][start:], label=dicc_v2['leyenda'])
    plt.xlabel(rotulos['eje x'])
    plt.ylabel(rotulos['eje y'])
    plt.title(rotulos['titulo'])
    plt.legend()
    plt.show()
    plt.close()
    
def plotDistanciaVectores(v1,v2, rotulos, start=10):
    if len(v1) != len(v2):
        print("Parametros invalidos")
        return
    distancia = np.abs(v1 - v2)
    n = np.arange(start,len(v1))
    plt.plot(n,distancia[start:])
    plt.xlabel(rotulos['eje x'])
    plt.ylabel(rotulos['eje y'])
    plt.title(rotulos['titulo'])
    plt.show()
    plt.close()
```

```python
 #Definimos diccionario de funiones
funciones = {'inv' : inversa,
             'solve' : resolverLU, 'norma' : norma2Vectorial}
v1,v2 = perfomance(funciones,rango_de_iteracion=(10,200))
dv1 = {'vector': v1, 'leyenda': "Resolver (LU)x=b"}
dv2 = {'vector': v2, 'leyenda': "x = inversa(A)@b"}
rotulos = {'titulo' : 'Errores relativos $\|Ax-b\|_2/\|b\|_2$',
           'eje x' : 'Dimensión de matriz',
           'eje y' : '$\sum_{i = 0}^{10}{ln(\|A_i*x_i-b_i\|_2/\|b_i\|_2)}$'}
    
plotearVectores(dv1, dv2, rotulos)
```

### Conclusiónes Experimento 1

* El método de descomponsición PA=LU para resolver sistemas de ecuaciones lineales tiende a producir soluciones más estables en comparacion a resolverlo con la inversa, esto se debe a la diferencia de presicion en ambos métodos. Ádemas, como podemos observar en el grafico 2 (distancia de errores relativos) la presición de la inversa empeora con respecto a la descomposición PA=LU a medida aumenta el tamaño del sistema lineal.

# Conclusiones finales

Los gráficos generados son curvas no suaves. Podemos atribuir esta naturaleza a los siguientes factores:
* Dado que son generados de forma estocástica, no se logra uniformidad en los valores asociados a las matrices (número de condición).
* Aunque utilizamos la precisión de máquina estándar y considerando que los valores $a$ tanto de matrices como de vectores están en un rango muy acotado, es decir, $a \in [-1,1)$, cuanto más pequeño sea el intervalo...

Todos los experimentos coinciden en:

* A la hora de resolver sistemas lineales, el método con mejor rendimiento es la descomposición PA=LU en comparación con el método de la inversa. Este comportamiento es independiente de la precisión de máquina seleccionada y de la función que se utilice para representar dicho método, ya sea las funciones desarrolladas en este trabajo o las obtenidas de la librería numpy/linalg.

En resumen, podemos concluir que al resolver sistemas lineales con métodos numéricos y ejecutados por una máquina, el resultado siempre estará propenso a errores. Por lo tanto, solo podremos obtener aproximaciones. No obstante, para obtener mejores resultados, es importante evaluar qué método nos ofrece un mejor rendimiento, el tipo de precisión numérica, ya que al tener mayor espacio de memoria por valor numérico para su cálculo, la aproximación tendrá un menor error y, por lo tanto, será más precisa. Por último, pero no menos importante, es fundamental comprender en profundidad cómo estos errores se propagan al diseñar algoritmos más estables para la resolución de cualquier problema.
